```python

from sklearn.preprocessing import MinMaxScaler
datatypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
normarization = data.select_dtypes(include=datatypes)
for col in normarization.columns:
    MinMaxScaler(col)
    #minmaxnorm =(x-min(x)/(max(x)-min(x)))- x=column of data
    #normalise columns having int and float datatype
```

```python
#K means algo- clustering of datapoints
#specify no of clusters, eg 10
#initialize the centroid
#select K datapoints
#calculate distance, and check to which cluster it is closest to
#songs from the different genres could have quite similar characteristics, and that's not fine.
#create a new feature, which would differ the songs from different groups
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
features = kmeans.fit_predict(normarization)
data['features'] = features
MinMaxScaler(data['features'])
#differentiates songs from different categories using K means
```

```python
#choose the songs with the smallest distances
class Spotify_Recommendation():
    def __init__(self, dataset):#self represents the object of the class itself
        self.dataset = dataset
        ##function which returns recommendations, we can also choose the amount of songs to be recommended
    def recommend(self, songs, amount=1):
        distance = []
        #choosing the data for our song
        song = self.dataset[(self.dataset.name.str.lower() == songs.lower())].head(1).values[0]# song is filtered
        #dropping the data with our song
        rec = self.dataset[self.dataset.name.str.lower() != songs.lower()]
        for songs in tqdm(rec.values):
            d = 0
            for col in np.arange(len(rec.columns)):
               #indeces of non-numerical columns
                if not col in [1, 6, 12, 14, 18]:
                    #calculating the manhettan distances for each numerical feature
                    d = d + np.absolute(float(song[col]) - float(songs[col]))
            distance.append(d)
            #sorting our data to be ascending by 'distance' feature
        rec['distance'] = distance
        rec = rec.sort_values('distance')
        columns = ['artists', 'name']
        return rec[columns][:amount]
#the object of our SpotifyRecommender.
recommendations = Spotify_Recommendation(data)
#recommendations.recommend("Lovers Rock", 10)
recommendations.recommend("The Monster", 10)
```
